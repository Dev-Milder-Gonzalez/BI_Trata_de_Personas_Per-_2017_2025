### 4.10. Creamos la columna ```"A칌O_MES"```

````python
# Creamos la columna A칌O_MES como string (No usamos un DateTime)
df["A칌O_MES"] = df["A칌O"].astype(str) + '-' + df["MES"].astype(str).str.zfill(2)
v = "A칌O_MES" in df.columns
print(f'Columnas A칌O_MES como columna del dataframe: {v}')
````

### 4.11. Reordenamos lso datos por columnas 

````python

orden_columnas = ["A칌O", "MES", "A칌O_MES", "CCDD", "DEPARTAMENTO", "PROVINCIA", "DISTRITO", "COMISARIA", "CATEGORIA", "SUBCATEGORIA", "VALOR"]
df = df[orden_columnas]
print(df.columns)
````

### 4.12. Ordenamos registros o filas jer치rquicamente

````python

df = df.sort_values(["A칌O", "MES", "A칌O_MES", "CCDD", "DEPARTAMENTO", "PROVINCIA", "DISTRITO", "COMISARIA", "CATEGORIA", "SUBCATEGORIA", "VALOR"]).reset_index(drop = True)
print("Datos ordenados perfectamente por jerarqu칤a")
print(df.head(86))
````

## 5. Pivot y variables九

````python
index_cols = ["A칌O", "MES", "A칌O_MES", "CCDD", "DEPARTAMENTO", "PROVINCIA", "DISTRITO", "COMISARIA"]
df_wide = df.pivot_table(index=index_cols, columns=["CATEGORIA", "SUBCATEGORIA"], values="VALOR", aggfunc="sum", fill_value=0).reset_index()
print(f'Dimensiones:\n{df_wide.shape}')
print(f'DataFrame ancho:\n{df_wide.head(10)}')
````

## 6. Normalizar columnas九꽲잺
### 6.1. Normalizamos nombres驕놾잺 

````python
df_wide.columns = [
    f"{cat}_{sub}".upper()
    .replace(" ", "_").replace("/", "_").replace("(", "_").replace(")", "_")
    .replace("츼", "A").replace("칄", "E").replace("칈", "I").replace("칍", "O").replace("칔", "U")
    .strip('_').strip()
    if sub else cat 
    for cat, sub in df_wide.columns
    ]
print(f'Dimensiones: {df_wide.shape }')
print(f'DataFrame sin MultiIndex:\n{df_wide.head(10)}')
````

### 6.2. Creamos columna: columna para contar los menores de 18 a침os游 

````python
menores_cols = [
    col
    for col in df_wide.columns 
    if re.search(r"(0_A_5|6_A_11|12_A_11)", col)]
df_wide["MENORES_18"] = df_wide[menores_cols].sum(axis=1)
````

## 7. DataSet Maestro游뉦잺

````python
df_wide.to_excel('../data/processed/dataset_maestro_wide.xlsx', index=False)
print("ETL completado - Maestro guardado")
print(f'DataSet maestro:\n{df_wide.head(10)}')
````
