## 5. Pivot y variables九

````python
index_cols = ['A칌O', 'MES', 'A칌O_MES', 'CCDD', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'COMISARIA']
df_wide = df.pivot_table(index=index_cols, columns=['CATEGORIA', 'SUBCATEGORIA'], values='VALOR', aggfunc='sum', fill_value=0).reset_index()
print(f'DataFrame ancho:\n{df_wide}')
````

````python
df_wide.columns = [f"{cat}_{sub}".replace(' ', '_').upper() if sub else cat for cat, sub in df_wide.columns]
````

````python
df_wide['TOTAL_CASOS'] = df_wide.get('CAPTACION_C_TOTAL', 0)
df_wide['TOTAL_EXPLOTACION'] = df_wide.get('EXPLOTACION_F2_TOTAL', 0)
df_wide['EXPLOTACION_SEXUAL'] = df_wide.get('FINALIDAD_DEL_TIPO_DE_DELITO_F1_EXPLOTACION_SEXUAL', 0)

menores_cols = [col for col in df_wide.columns if '0_A_5' in col or '6_A_11' in col or '12_A_17' in col]
df_wide['MENORES_18'] = df_wide[menores_cols].sum(axis=1)

df_wide.to_excel('../data/processed/dataset_maestro_wide.xlsx', index=False)

print("ETL completado - Maestro guardado")
df_wide.sample(10)
````
