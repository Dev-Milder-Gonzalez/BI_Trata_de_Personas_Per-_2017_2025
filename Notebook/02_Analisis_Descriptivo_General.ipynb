```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#ConfiguraciÃ³n presetnaciÃ³n de datos
pd.set_option("display.max_columns", None) # Mostrar todas las columnas
pd.set_option("display.max_rows", None) # Mostrar mÃ¡s filas
pd.set_option("display.width", None) # Evita saltos de lineas raros
pd.set_option("display.float_format", lambda x: f'{x:,.4f}') #Formato de nÃºmero
pd.set_option("display.expand_frame_repr", False)
```

## 1. Cargamos el DatSet Maestroâœ…

```python
df_wide = pd.read_excel("../data/processed/dataset_maestro_wide.xlsx")
print(f'âœ…DataSet maestro cargado')
print(f'Dimensiones: {df_wide.shape}')
```

## 2. Series de explotaciÃ³nðŸ“ˆ

### 2.1. Serie de tiempo anual

```python
anual_explotacion = df_wide.groupby("AÃ‘O")["EXPLOTACION_F2_TOTAL"].sum()
print(type(anual_explotacion))
print('-' * 70)
print(f'ExplotaciÃ³n anual desde Enero 217 hasta Septiembre 2025, PerÃº:\n{anual_explotacion}')
```

```python
sns.lineplot(x = anual_explotacion.index, y = anual_explotacion, marker = "o",color = "red")
for i, v in enumerate(anual_explotacion):
    #Anadir valores sobre cada punto
    plt.text(anual_explotacion.index[i], v, str(int(v)), ha = "center")

#TÃ­tulo del grÃ¡fico
plt.title("ExplotaciÃ³n anual desde Enero 217 hasta Septiembre 2025, PerÃº")
plt.xlabel("AÃ±o")
plt.ylabel("Casos totales")
plt.grid(True)
plt.tight_layout()
plt.savefig("../reports/Series_tiempo/Victimas_de_explotacion_por_aÃ±o.png", dpi = 500)
plt.show()
```

### 2.2. Serie de tiempo por aÃ±o y mes

```python
explotacion_anio_mes = df_wide.groupby("AÃ‘O_MES")["EXPLOTACION_F2_TOTAL"].sum()
print(f"ExplotaciÃ³n por aÃ±o y mes desde Enero 217 hasta Septiembre 2025, PerÃº:\n{explotacion_anio_mes}")
```

```python
plt.figure(figsize=(22, 7))

sns.lineplot(x=explotacion_anio_mes.index, 
             y=explotacion_anio_mes.values, 
             color="red", 
             marker="o")

for x, y in zip(explotacion_anio_mes.index, explotacion_anio_mes.values):
    plt.text(x, y, str(int(y)), ha='center', fontsize=8)

plt.title("ExplotaciÃ³n por aÃ±o y mes desde Enero 2017 hasta Septiembre 2025, PerÃº")
plt.xlabel("AÃ±o-mes")
plt.ylabel("Casos")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.savefig("../reports/Series_tiempo/Victimas_de_explotacion_por_aÃ±o_mes.png", dpi = 800)
plt.show()
```

## 5. AnÃ¡lisis de las mujeres vÃ­ctimas ðŸ“ˆ 

### 5.1. Preparamos los datos

```python
print("EM_EF_TOTAL_EM_EF_TOTAL colo columna del df_wide:","EM_EF_TOTAL_EM_EF_TOTAL" in df_wide.columns)
total_victimas_aÃ±o = df_wide.groupby("AÃ‘O", sort=True)["EM_EF_TOTAL_EM_EF_TOTAL"].sum()
total_mujeres_aÃ±o = df_wide.groupby("AÃ‘O", sort=True)["FEMENINO_GRUPO_EDAD_EF_SUB_TOTAL"].sum() 
#Porcentaje de mujeres por registro
porc_mujeres_aÃ±o = (total_mujeres_aÃ±o/total_victimas_aÃ±o * 100).round(2)
print('NÃºmero total de vÃ­ctimas:',type(total_victimas_aÃ±o))
print(total_victimas_aÃ±o)
print('-' * 80)
print('NÃºmero total de muejeres victÃ­mas:', type(total_mujeres_aÃ±o))
print(total_mujeres_aÃ±o)
print('-' * 80)
print('Porcentaje de mujeres vÃ­ctimas:', type(porc_mujeres_aÃ±o))
print(porc_mujeres_aÃ±o)
```

### 5.2. Porcentaje de mujeres anual

```python
sns.lineplot(x=porc_mujeres_aÃ±o.index, y=porc_mujeres_aÃ±o.values, color = "#bbff3d")
for i ,v in enumerate(porc_mujeres_aÃ±o.values):
    plt.text(porc_mujeres_aÃ±o.index[i] , v, str(int(v)), ha="center", va = "bottom")

plt.title("Porcentaje de mujeres al aÃ±o vÃ­ctimas del delito de trata de personas entre Enero del 2017 y Septiembre del 2025")
plt.xlabel("AÃ±o")
plt.ylabel("Porcentaje(%)")
plt.ylim(0, 100)
plt.grid()
plt.tight_layout()
plt.savefig("../reports/Series_tiempo/Porcentaje_anual_mujeres_victimas.png")
plt.show()
```

### 5.3. Porcentaje de mujeres vÃ­ctimas en todo el periodo (Enero del 2017 y Septiembre del 2025)

```python
porc_muj_global = total_mujeres_aÃ±o.sum() / total_victimas_aÃ±o.sum() * 100
porc_var_global = 100 -porc_muj_global
print(f'Porcentaje de mujeres vÃ­ctimas: {porc_muj_global:.4f} %')
print(f'Porcentaje de mujeres vÃ­ctimas: {porc_var_global:.4f} %')
plt.pie(
    [porc_var_global, porc_muj_global], 
    labels= ["Hombres", "Mujeres"],
    autopct = "%.2f%%"
)
plt.title("Porcentaje de varones y mujeres vÃ­ctimas del delito de trata de personas entre Enero del 2017 y Septiembre del 2025")
plt.grid()
plt.tight_layout()
plt.savefig("../reports/Porcentaje_de_mujeres_varones_victimas_en_todo_el_periodo.png")
plt.show()
```
